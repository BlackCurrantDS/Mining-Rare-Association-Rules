```
import pandas as pd

df = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/heart-disease/processed.cleveland.data', 
    sep=',', 
    header=None,
    names = [
    'AGE'
    ,'SEX'
    ,'cp'
    ,'trestbps'
    ,'chol'
    ,'fbs'
    ,'restecg'
    ,'thalach'
    ,'exang'
    ,'oldpeak'
    ,'slope'
    ,'ca'
    ,'thal'
    ,'class'
]
)

print(df)
```

For rules generations

```
for i,n in enumerate(df.columns):
  if n=='class':
    df[n] = 'class@' + df[n].astype(str)
  else:
    df[n] = 'A'+str(i)+'@' + df[n].astype(str)
```

```
df['class'] = df['class'].str.replace('2', '1').str.replace('3', '1').str.replace('4', '1')
```

```
df
```

```
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, stratify = df['class'], test_size = .2)
```

```
train.to_csv("/content/drive/MyDrive/Clev_DataSet/clev_train.csv", header=None,index=False)
test.to_csv("/content/drive/MyDrive/Clev_DataSet/clev_test.csv", header=None,index=False)
```

For RP-growth Data

```
uni = pd.DataFrame()
for col in df:
    uni= pd.concat([uni, pd.DataFrame(df[col].unique())], ignore_index=True)
```

```
uni
```

```
my_dict = uni.to_dict()
```

```
my_dict
```

```
my_dict = my_dict[0]
```

```
new_dict = dict((value, key) for key, value in my_dict.items())
```

```
for col in train:
    train[col] = train[col].replace(new_dict)
```

```
train
```

```
for col in test:
    test[col] = test[col].replace(new_dict)
```
