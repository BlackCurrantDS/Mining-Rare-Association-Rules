```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = [['SVM', 'BreastCancer', .66, 'Clustering'], ['SVM', 'BreastCancer', .50, 'RPP'], ['SVM', 'BreastCancer', .50, 'RP-Growth']
        ,['RF', 'BreastCancer', .71, 'Clustering'], ['RF', 'BreastCancer', .50, 'RPP'], ['RF', 'BreastCancer', .50, 'RP-Growth']
        ,['MLP', 'BreastCancer', .66, 'Clustering'], ['MLP', 'BreastCancer', .50 , 'RPP'], ['MLP', 'BreastCancer', .50 , 'RP-Growth']
        ,['SVM', 'Cleveland', .84, 'Clustering'], ['SVM', 'Cleveland', .50, 'RPP'], ['SVM', 'Cleveland', .50, 'RP-Growth']
        ,['RF', 'Cleveland', .82, 'Clustering'], ['RF', 'Cleveland', .50, 'RPP'], ['RF', 'Cleveland', .50, 'RP-Growth']
        ,['MLP', 'Cleveland', .82, 'Clustering'], ['MLP', 'Cleveland', .50 , 'RPP'], ['MLP', 'Cleveland', .50 , 'RP-Growth']
        ,['SVM', 'Hepatitis', .7461, 'Clustering'], ['SVM', 'Hepatitis', .50, 'RPP'], ['SVM', 'Hepatitis', .50, 'RP-Growth']
        ,['RF', 'Hepatitis', .744, 'Clustering'], ['RF', 'Hepatitis', .50, 'RPP'], ['RF', 'Hepatitis', .50, 'RP-Growth']
        ,['MLP', 'Hepatitis', 0.757, 'Clustering'], ['MLP', 'Hepatitis', .50 , 'RPP'], ['MLP', 'Hepatitis', .50 , 'RP-Growth']
        ]
```

```
df = pd.DataFrame(data, columns = ['Classifier', 'DataSet', 'Score', 'Method']) 
```

```
ax = sns.barplot(x="Classifier", y="Score", hue="Method", data=df)
```

```
sns.catplot(x="Classifier", y="Score",
                hue="Method", col="DataSet",
                data=df, kind="bar",
                height=4, aspect=.7)
plt.xlabel("Classifier")
#plt.ylabel("F1-Score")
#plt.title("F1 scores") # You can comment this line out if you don't need title
plt.savefig("Paper_F1_Score.png",bbox_inches='tight')
```
