```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
data = [['SVM', 'BreastCancer', .73, 'Clustering'], ['SVM', 'BreastCancer', .76, 'RPP'], ['SVM', 'BreastCancer', .76, 'RP-Growth']
        ,['RF', 'BreastCancer', .75, 'Clustering'], ['RF', 'BreastCancer', .77, 'RPP'], ['RF', 'BreastCancer', .77, 'RP-Growth']
        ,['MLP', 'BreastCancer', .77, 'Clustering'], ['MLP', 'BreastCancer', .82 , 'RPP'], ['MLP', 'BreastCancer', .82 , 'RP-Growth']
        ,['SVM', 'Cleveland', .84, 'Clustering'], ['SVM', 'Cleveland', .87, 'RPP'], ['SVM', 'Cleveland', .87, 'RP-Growth']
        ,['RF', 'Cleveland', .82, 'Clustering'], ['RF', 'Cleveland', .88, 'RPP'], ['RF', 'Cleveland', .88, 'RP-Growth']
        ,['MLP', 'Cleveland', .82, 'Clustering'], ['MLP', 'Cleveland', .86 , 'RPP'], ['MLP', 'Cleveland', .86 , 'RP-Growth']
        ,['SVM', 'Hepatitis', .7461, 'Clustering'], ['SVM', 'Hepatitis', .78, 'RPP'], ['SVM', 'Hepatitis', .78, 'RP-Growth']
        ,['RF', 'Hepatitis', .744, 'Clustering'], ['RF', 'Hepatitis', .78, 'RPP'], ['RF', 'Hepatitis', .78, 'RP-Growth']
        ,['MLP', 'Hepatitis', 0.757, 'Clustering'], ['MLP', 'Hepatitis', .78 , 'RPP'], ['MLP', 'Hepatitis', .78 , 'RP-Growth']
        ]
```

```
df = pd.DataFrame(data, columns = ['Classifier', 'DataSet', 'F1-Score', 'Method']) 
```

```
ax = sns.barplot(x="Classifier", y="F1-Score", hue="Method", data=df)
```

```
sns.catplot(x="Classifier", y="F1-Score",
                hue="Method", col="DataSet",
                data=df, kind="bar",
                height=4, aspect=.7)
plt.xlabel("Classifier")
#plt.ylabel("F1-Score")
#plt.title("F1 scores") # You can comment this line out if you don't need title
plt.savefig("Paper_F1_Score.png",bbox_inches='tight')
```
