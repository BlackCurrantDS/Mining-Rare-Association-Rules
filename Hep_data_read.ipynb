```
import pandas as pd

df = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/hepatitis/hepatitis.data', 
    sep=',', 
    header=None,
    names = ['Class'
    ,'AGE'
    ,'SEX'
    ,'STEROID'
    ,'ANTIVIRALS'
    ,'FATIGUE'
    ,'MALAISE'
    ,'ANOREXIA'
    ,'LIVER BIG'
    ,'LIVER FIRM'
    ,'SPLEEN PALPABLE'
    ,'SPIDERS'
    ,'ASCITES'
    ,'VARICES'
    ,'BILIRUBIN'
    ,'ALK PHOSPHATE'
    ,'SGOT'
    ,'ALBUMIN'
    ,'PROTIME'
    ,'HISTOLOGY']
)

print(df)
```

```
for i,n in enumerate(df.columns):
  #print(i,n)
  df[n] = 'A'+str(i)+'@' + df[n].astype(str)
```

```
df
```

```
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, stratify = df['Class'], test_size = .2)
```

```
train.to_csv("/content/drive/MyDrive/Hep_DataSet/hep_train.csv")
test.to_csv("/content/drive/MyDrive/Hep_DataSet/hep_test.csv")
```

For RP-growth Data

```
uni = pd.DataFrame()
for col in df:
    uni= pd.concat([uni, pd.DataFrame(df[col].unique())], ignore_index=True)
```

```
uni
```

```
my_dict = uni.to_dict()
```

```
my_dict
```

```
my_dict = my_dict[0]
```

```
new_dict = dict((value, key) for key, value in my_dict.items())
```

```
for col in train:
    train[col] = train[col].replace(new_dict)
```

```
train
```

```
for col in test:
    test[col] = test[col].replace(new_dict)
```

```
train.to_csv("/content/drive/MyDrive/Hep_DataSet/hep_rp_train.csv")
test.to_csv("/content/drive/MyDrive/Hep_DataSet/hep_rp_test.csv")
```
