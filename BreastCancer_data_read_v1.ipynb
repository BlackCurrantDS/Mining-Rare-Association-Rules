Attribute information: 
1. Class: no-recurrence-events, recurrence-events
2. age: 10-19, 20-29, 30-39, 40-49, 50-59, 60-69, 70-79, 80-89, 90-99.
3. menopause: lt40, ge40, premeno.
4. tumor-size: 0-4, 5-9, 10-14, 15-19, 20-24, 25-29, 30-34, 35-39, 40-44, 45-49, 50-54, 55-59.
5. inv-nodes: 0-2, 3-5, 6-8, 9-11, 12-14, 15-17, 18-20, 21-23, 24-26, 27-29, 30-32, 33-35, 36-39.
6. node-caps: yes, no.
7. deg-malig: 1, 2, 3.
8. breast: left, right.
9. breast-quad: left-up, left-low, right-up, right-low, central.
10. irradiat: yes, no.

```
import pandas as pd

df = pd.read_csv(
    'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer/breast-cancer.data', 
    sep=',', 
    header=None,
    names = ['class'
    ,'AGE'
    ,'menopause'
    ,'tumor-size'
    ,'inv-nodes'
    ,'node-caps'
    ,'deg-malig'
    ,'breast'
    ,'breast-quad'
    ,'irradiat']
  
)

print(df)
```

Brining back to original

Converting unexpected rules to original attributes

```
mapping = {
    
     'A10':'SPLEEN PALPABLE',
    'A11':'SPIDERS',
    'A12':'ASCITES',
    'A13':'VARICES',
    'A14':'ALK PHOSPHATE',
    'A15':'SGOT',
    'A16':'ALBUMIN',
    'A17':'PROTIME',
    'A18':'HISTOLOGY',
    
	'A0':'class'
    'A1':'AGE',
    'A2':'SEX',
    'A3':'STEROID',
    'A4':'ANTIVIRALS',
    'A5':'FATIGUE',
    'A6':'MALAISE',
    'A7':'ANOREXIA',
    'A8':'LIVER BIG',
    'A9':'LIVER FIRM'
   
           
           }
```

```
import fileinput

text = "/content/drive/MyDrive/Clev_DataSet/RPGrowth/unexpected_rule_file_json.txt"
fields = mapping

lst = []
for line in fileinput.input(text):
    for field in fields:
        if field in line:
            line = line.replace(field, fields[field])
    lst.append(line)

print(lst)
```

```
f = open('/content/drive/MyDrive/Clev_DataSet/RPGrowth/unexpected_rule_file_mapped.txt','w')
for line in lst:
    f.write(line)
f.close()
```
