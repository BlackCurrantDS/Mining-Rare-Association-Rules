```
import pyfpgrowth
```

```
#demo example
transactions = [[1, 2, 5],
                [2, 4],
                [2, 3],
                [1, 2, 4],
                [1, 3],
                [2, 3],
                [1, 3],
                [1, 2, 3, 5],
                [1, 2, 3]]
```

```
patterns = pyfpgrowth.find_frequent_patterns(transactions, 2)
```

```
patterns
```

```
rules = pyfpgrowth.generate_association_rules(patterns, 0.7)
```

```
rules
```

```
import pandas as pd
df = pd.read_csv('/content/breast_train_transactions.txt', header = None, sep="," , names=['a','b','c', 'd', 'e', 'f','g', 'h', 'i', 'k'])
```

```
df.head(5)
```

```
trans = df.values.tolist()
```

```
patterns = pyfpgrowth.find_frequent_patterns(trans, 8)
```

```
patterns
```

```
rules = pyfpgrowth.generate_association_rules(patterns, 0.8)
rules
```

```
type(rules)
```

```
with open('fpgrowth_out.txt', 'w') as f:
    print(rules, file=f)
```

```
#remove all rules wihtout class
refined_rules = {k: v for k, v in rules.items() if 'class@no' in v}
```

```
refined_rules
```

```
for key, value in rules.items():
  print(key)
  print(value)
  print(len(value[0]))
  print(value[0])
```

```
for key, value in rules.items():
  if ('class@no' or  'class@yes') not in key and len(value[0])==1 and "class@yes" in value[0] : 
    print(key)
    print(value)
```
