Converting unexpected rules to original attributes

```
mapping = {
    
     'A10':'SPLEEN PALPABLE',
    'A11':'SPIDERS',
    'A12':'ASCITES',
    'A13':'VARICES',
    'A14':'ALK PHOSPHATE',
    'A15':'SGOT',
    'A16':'ALBUMIN',
    'A17':'PROTIME',
    'A18':'HISTOLOGY',
    
	'A0':'class',
    'A1':'AGE',
    'A2':'SEX',
    'A3':'STEROID',
    'A4':'ANTIVIRALS',
    'A5':'FATIGUE',
    'A6':'MALAISE',
    'A7':'ANOREXIA',
    'A8':'LIVER BIG',
    'A9':'LIVER FIRM'
   
           
           }
```

```
import fileinput

text = "/content/hep_rpgrwoth_rules.csv"
fields = mapping

lst = []
for line in fileinput.input(text):
    for field in fields:
        if field in line:
            line = line.replace(field, fields[field])
    lst.append(line)

print(lst)
```

```
f = open('/content/hep_rpgrwoth_rules.csv','w')
for line in lst:
    f.write(line)
f.close()
```
